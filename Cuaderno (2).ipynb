
###Factorial
Codigo


```
def factorial(num):
    if isinstance(num, int):
        resultado = 1
        if num == 0:
            return resultado
        if num > 0:
            for numero in range (1, num + 1):
                resultado = resultado * numero
            return resultado
        else:
            return "Error: El numero tiene que ser mayor que cero"
    else:
        return "Error: No son numero"
```

```
class Circulo:
    def __init__(self, radio):
        self.radioCirculo = radio
        self.pi = 3.14

    def calcularArea(self):
        resultado = 3.14 * (self.radioCirculo ** 2)
        return resultado
    def calcularPerimetro(self):
        resultado = 2 * self.pi * self.radioCirculo
        return resultado

class rectangulo:
    def __init__(self, alto, ancho):
        self.alto = alto
        self.ancho = ancho

    def area(self):
        resultado = self.alto * self.ancho
        return resultado

    def perimetro(self):
        resultado = 2*self.ancho + 2*self.alto
        return resultado

    def mostrarAncho(self):
        ancho = self.ancho
        return ancho
```

```
class calculadora:
    def __init__(self, operador1, operador2):
        self.op1 = operador1
        self.op2 = operador2

    def suma(self):
        suma = self.op1 + self.op2
        return suma

    def resta(self):
        resultado = self.op1 - self.op2
        return resultado

    def multiplicacion(self):
        resultado = self.op1 * self.op2
        return resultado

    def division(self):
        resultado = self.op1 / self.op2
        return resultado

    def potencia(self):
        resultado = self.op1**self.op2
        return resultado

```
